Trabalho de conclusão de curso: Um modelo de Rede Neural de classificação para predição da tendência do preço Bitcoin

Eu usei Python, redes neurais LSTM e conhecimento de economia para construção de um modelo de classificação da tendência do preço do Bitcoin utilizando dados relacionados ao preçoe o volume de negociações. Os resultados obtidos foram satisfatórios, o modelo obteve uma acurácia de 55% e um retorno de investimento de 240%.
As ferramentas utilizadas foram:
Python, Pandas e Matplotlib 
Jupyter Notebook
Algorítimo de Machine Learning de Classificação
Redes Neurais LSTM
